```
import matplotlib.pyplot as plt
```

```
plt.hist(np.random.rand(100000), density=True, bins=200, histtype="step", color="blue", label="rand")
plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc= "upper left")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```
from mpl_toolkits.mplot3d import Axes3D
```

```
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
x, y = np.meshgrid(x, y)
R = np.sqrt(x**2 + y**2)
Z = np.sin(R)

figure = plt.figure(1, figsize= (12, 4))
subplot = plt.subplot(111, projection='3d')
surface = subplot.plot_surface(x, y, Z, rstride=1, cstride=1, cmap=plt.cm.coolwarm, linewidth=0.1)
plt.show
```
